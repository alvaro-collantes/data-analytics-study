**Week 05: Cleaning**

```python
#Libraries
import pandas as pd
#Load data
df = pd.read_csv("C:\ARCHIVOS\DOCUMENTS\Alvaro Universidad\M2 FIT\Practice\Sample - Superstore.csv",encoding = 'latin1')
```

```python
df.columns
```

**Exercise 1:Check and fill nulls**

```python
#Create a copy
df_clean = df.copy()
```

```python
#Check nulls 
df_clean.isnull().sum()[df_clean.isnull().sum()>0]
```

```python
#Fill null string values with unknown       
df_clean['Postal Code'] = df_clean['Postal Code'].fillna('Unknown')
```

```python
#Fill null numeric values with the median (more robust)
df_clean['Sales'] = df_clean['Sales'].fillna(df_clean['Sales'].median())
```

```python
#Drop columns with null values 
df_clean = df_clean.dropna(subset=['Order ID', 'Customer ID'])
```

**Exercise 2:Fix data types and clean text**

```python
# Convert date columns from text to datetime
df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'])
df_clean['Ship Date']  = pd.to_datetime(df_clean['Ship Date'])
```

```python
# Treat whitespace and normalize case
df_clean['Customer Name'] = df_clean['Customer Name'].str.strip()
df_clean['Category']      = df_clean['Category'].str.strip().str.title()
```

```python
# Verify
df_clean[['Order Date','Ship Date','Category']].dtypes
```

**Exercise 3:Remove duplicates**

```python
print(f'Before: {len(df_clean)}')
print(f'Exact duplicates: {df_clean.duplicated().sum()}')

df_clean = df_clean.drop_duplicates()

# Deduplicate by Order ID (keep first occurrence per order)
df_unique_orders = df_clean.drop_duplicates(subset='Order ID', keep='first')
print(f'Unique orders: {len(df_unique_orders)}')
```
