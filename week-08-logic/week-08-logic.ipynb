```python
#Libraries
import pandas as pd
```

```python
#Load data
df = pd.read_csv("C:\ARCHIVOS\DOCUMENTS\Alvaro Universidad\M2 FIT\Practice\Sample - Superstore.csv",encoding = 'latin1')
```

```python
def classify_margin(profit,sales):
    if sales == 0:
        return 'No Data'
    margin = profit/sales
    if margin >= 0.25: return 'High'
    if margin >= 0.10: return 'Medium'
    if margin >= 0: return 'Low'
    else: return 'Loss'

df['margin_tier']=df.apply(
    lambda row: classify_margin(row['Profit'],row['Sales']),
    axis =1 
)
df['margin_tier'].value_counts()
```
