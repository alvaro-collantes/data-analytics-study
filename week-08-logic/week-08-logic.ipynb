**Exercise 1: Classification**

```python
def classify_margin(profit,sales):
    if sales == 0:
        return 'No Data'
    margin = profit/sales
    if margin >= 0.25: return 'High'
    if margin >= 0.10: return 'Medium'
    if margin >= 0: return 'Low'
    else: return 'Loss'

df['margin_tier']=df.apply(
    lambda row: classify_margin(row['Profit'],row['Sales']),
    axis =1 
)
df['margin_tier'].value_counts()
```

**Exercise 2: List comprehensionn**

```python
#Lambda function
df['Sales_K'] = df['Sales'].apply(lambda x: round(x/1000,1))

#Create list comprehension 
clean_cat = [c.strip().upper() for c in df['Category'].unique()]
print(clean_cat)

#Condition
high_sales = [v for v in df['Sales'] if v > 1000]
print(f'sales above $1000: {len(high_sales)}')
```

**Exercise 3: Analysis Function**

```python
#Create a function that returns a grouped summary with total, mean, count and margin
def summarize_by(df,group_col,value_col ='Sales'):
    return(df.groupby(group_col)
            .agg(total=(value_col,'sum'),
                 avg =(value_col,'mean'),
                 count = (value_col,'count'),
                 profit = ('Profit','sum'))
            .assign(margin_pct = lambda x: round(x['profit']/x['total']*100,2))
            .sort_values('total', ascending = False))

#Reuse for any column
display(summarize_by(df,'Category'))
display(summarize_by(df,'Region'))
display(summarize_by(df,'Segment'))
```
