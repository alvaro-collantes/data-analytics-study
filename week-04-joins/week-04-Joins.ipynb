```python
#Libraries
import pandas as pd
#Load data
df = pd.read_csv("C:\ARCHIVOS\DOCUMENTS\Alvaro Universidad\M2 FIT\Practice\Sample - Superstore.csv",encoding = 'latin1')
```

```python
df.columns
```

```python
#Exercise 0: Prepare 2 datasets 
#Dataset 1
orders = df[['Order ID','Order Date','Customer ID','Region','Segment']].drop_duplicates()
#Dataset 2
line_items = df[['Order ID','Product Name','Category','Sales','Profit']]
print(f'Orders: {orders.shape} | Line items: {line_items.shape}')
```

```python
#Exercise 1: Join 2 tables
combined = pd.merge(orders,line_items, on = 'Order ID', how = 'inner')
print(f'Result: {combined.shape}')
```

```python
#Exercise 2: Use of Agg functions
result = combined.groupby('Segment').agg(
    num_orders = ('Order ID','nunique'),
    total_sales = ('Sales','sum'),
    avg_ticket = ('Sales','mean')
).round(2).sort_values('total_sales', ascending=False)
result
```

```python
#Exercise 3: Find unmatched rows

#Add fake data
fake = pd.DataFrame({'Order ID': ['CA-FAKE-001'], 'Order Date':['2017-06-15'],'Customer ID':['CUST-999'],'Region':['West'],'Segment':['Consumer']})
orders_extra = pd.concat([orders,fake],ignore_index = True)

#Left Join
left = pd.merge(orders_extra,line_items, on='Order ID', how='left')
unmatched = left[left['Sales'].isna()]
unmatched
```
