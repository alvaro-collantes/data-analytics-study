**Week 2: Filters**

*1.Basic Filter*

```python
#Filter rows by Region , West
#Create serie
west = df[df['Region']=='West']
#Display total number & df
print(f"Total West Rows: {len(west)}")
display(west.head())
```

*2.Filter with 2 Conditions*

```python
#Filter by west region where the sales are higher than 500
#Use and operator (&)
#It needs to use 2 parenthesis

#Create serie
filter2 = df[(df['Region']=='West') & (df['Sales']>500)]

#Display 
print(f"Total rows: {len(filter2)}")
display(filter2.head())
```

*Other Way*

```python
#Using query gives the same result
#way 1
otherway1 = df.query("Region == 'West' and Sales > 500")
#way 2
limit = 500
otherway2 = df.query("Region == 'West' and Sales > @limit")

#print
print(f"Result 1: {len(otherway1)},Result 2: {len(otherway2)}")
```

*3.Filter with 1 or 2 Conditions*

```python
#Show rows where the category is Technology or Furniture
#Use Or operator (|) or .isin()

#Create df
filter3 = df[df['Category'].isin(['Technology','Furniture'])]

#Display
print(f"Total Rows: {len(filter3)}")
display(filter3.head())
```

*4.Filter text*

```python
#Filter chairs by product name

#Create df
chairs = df[df['Product Name'].str.contains('Chair', case = False, na = False)]
#str.contains filter by text
#case=False dont make distinction between capitl and lower case words
#na = false treats all na as false

#Display
print(f"Products with Chair in its names: {len(chairs)}")
#First 10
print(f"First 10 Products: \n {chairs['Product Name'].unique()[:10]}")
```

***Filter by Ship Mode: Standard class and sales higher than 600. With the 2 ways***

```python
#Get Ship Mode names and quantity 
ship_modes_n = df['Ship Mode'].unique()
ship_modes_quantity = df['Ship Mode'].nunique()
print(f"Ship Modes Names: {ship_modes_n}")
print(f"Ship Modes Quantity: {ship_modes_quantity}")
```

```python
#Way 1 (There are spaces, use backstrings)
filter_5 = df.query("`Ship Mode` == 'Second Class' & Sales > 600")
display(filter_5.head())
print(f"Way 1: {len(filter_5)}")
```

```python
#Way 2
ship_modes = 'Second Class'
sales_values = 600
filter_6 = df.query("`Ship Mode` == @ship_modes and Sales >@sales_values")
display(filter_6.head())
print(f"Way 2: {len(filter_6)}")
```

```python
#Comparison
print(f"Way 1: {len(filter_5)}, Way 2:{len(filter_6)}")
```
