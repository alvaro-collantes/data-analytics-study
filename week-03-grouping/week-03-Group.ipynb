```python
#Libraries
import pandas as pd
```

```python
#Load data
df = pd.read_csv("C:\ARCHIVOS\DOCUMENTS\Alvaro Universidad\M2 FIT\Practice\Sample - Superstore.csv",encoding = 'latin1')
```

```python
#Exercise 1: Total Sales per Category
result = df.groupby('Category')['Sales'].sum().round(2)
print(f"Total Sales per Category \n{result}")
```

```python
#Exercise 2: Multiple KPIs with Aggregate Functions
result2 = df.groupby('Category').agg(
    Total_Sales = ('Sales', 'sum'),
    Average_Sales = ('Sales', 'mean'),
    Total_Transactions = ('Sales','count'),
    Total_Profit = ('Profit','sum')
).round(2)
result2['Margin pct'] = round(result2['Total_Profit']/result2['Total_Sales']*100,2)
result2.sort_values('Total_Sales', ascending=False)
```

```python
#Exercise 3: Group by 2 columns + pivot table 
#Group by 2 columns
by_region_cat = df.groupby(['Region','Category'])['Sales'].sum().reset_index()
#Pivot Table
table = df.pivot_table(values='Sales', index='Region',columns='Category', aggfunc='sum')
round(table,0) 
```
