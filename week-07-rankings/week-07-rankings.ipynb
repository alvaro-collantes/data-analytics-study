```python
#Libraries
import pandas as pd
```

```python
#Load data
df = pd.read_csv("C:\ARCHIVOS\DOCUMENTS\Alvaro Universidad\M2 FIT\Practice\Sample - Superstore.csv",encoding = 'latin1')
```

**Exercise 1: Top N Products**

```python
top5 = (df.groupby('Product Name')['Sales']
        .sum()
        .sort_values(ascending=True)
        .head(5)
        .reset_index())
top5.columns = ['Products','Total Sales']
display(top5)
```

**Exercise 2: Rank Customers within each region**

```python
by_cust = df.groupby(['Region','Customer Name'])['Sales'].sum().reset_index()
by_cust['rank'] = (by_cust.groupby(['Region'])['Sales']
                   .rank(ascending=False, method='dense'))
#Top 3 per Region
top3 = by_cust[by_cust['rank']<=3].sort_values(['Region','rank'])
display(top3)
```
